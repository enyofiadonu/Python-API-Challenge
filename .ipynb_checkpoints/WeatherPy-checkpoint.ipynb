**Observations Based On The Data**
There is a clear relationship between temperature and latitude: as one gets further away from the equator, the more like you are to experience decreasing temperature.
There doesn't seem to be any relationship between cloudiness, windspeed and latitude. This could be because latitudes are horizontal bands across the planet which can cross through varying different climates, one latitude line could have cities with no clouds or cities with alot of clouds. 
There are common 'bands' of cloudiness (or common cloudiness) across all latitudes. 0 and 100 cloudiness are very common. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time 
from scipy.stats import linregress

from api_keys import weather_api_key 

from citipy import citipy

output_data_file = 'output_data/cities.csv'

lat_range = (-90,90)
lng_range = (-180,180)
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(lat_range[0],lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0],lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
query_units = 'imperial'
query_url = f"{url}appid={weather_api_key}&units={query_units}&q="
```

```python
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
city_new = []
count = 0
set_1 = 1
for city in cities:
    try:
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        wind_speed.append(response['wind']['speed'])
        if count > 48:
            count = 1 
            set_1 += 1
            city_new.append(city)
        else:
            count += 1
            city_new.append(city)
        print(f'Processing Record {count} of Set {set_1} | {city}')
    except:
        print(f'City not Found. Skipping....')
```

```python
weather_dict = {
    'City': city_new,
    'Lat': lat,
    'Lng': lng,
    'Max Temp': max_temp,
    'Humidity': humidity,
    'Cloudiness': cloudiness,
    'Wind Speed': wind_speed,
    'Country': country,
    'Date': date
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
weather_data.describe()
```

```python
weather_data.to_csv(output_data_file)
```

```python
weather_data.plot.scatter(x ='Lat',y='Max Temp', title = 'Temperature (F) vs. Latitude',edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.grid(b=True,which='major',axis='both',color='lightgrey')
plt.savefig('Figures/fig1.png')
plt.show()
```

This plot is showing how there is a max temperature of the planet around the equator, and steadily decreases as the latitude increase in the northern or southern direction.

```python
weather_data.plot.scatter(x='Lat',y='Humidity', title='Humidity (%) vs. Latitude',edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.grid(b=True,which='major',axis='both',color='lightgrey')
plt.savefig('Figures/fig2.png')
plt.show()
```

This plot shows the humidity of cities verses the latitude of those cities. There may be no correlation between latitude and humidity: most of these latitudes have a wide range of humidity.

```python
weather_data.plot.scatter(x='Lat',y='Cloudiness',title='Cloudiness (%) vs. Latitude',edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.grid(b=True,which='major',axis='both',color='lightgrey')
plt.savefig('Figures/fig3.png')
plt.show()
```

This plot shows the cloudiness of cities verses the latitude of those cities. This plot suggests there is no correlation between cloudiness and latitude, but there are common cloudiness (such as 0 % and 100%) across all latitudes.

```python
weather_data.plot.scatter(x='Lat',y='Wind Speed',title='Wind Speed (mph) vs. Latitude',edgecolors='black')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.grid(b=True,which='major',axis='both',color='lightgrey')
plt.savefig('Figures/fig4.png')
plt.show()
```

This plot shows the wind speed of cities verses the latitude of those cities. This plot suggests there is no correlation between latitude and wind speed, however slower wind speeds are more common across all latitudes. 

```python
northern_hemi_df = weather_data[weather_data['Lat'] >= 0].reset_index()
southern_hemi_df = weather_data[weather_data['Lat'] < 0].reset_index()
```

```python
x_values = northern_hemi_df['Lat']
y_values = northern_hemi_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values,edgecolors='black')
plt.title('Temperature (F) vs. Latitude of the Northern Hemispere')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.plot(x_values,regress_values,"m-",label=line_eq)
plt.legend()
plt.savefig('Figures/fig5.png')
plt.show()
```

This linear regression suggests there is a negative correlation between temperature and the latitudes of the northern hemishpere. As the latitude increases, the temperature tends to decrease

This linear regression suggests there is a sight positive correlation between humidity and the latitudes of the northern hemishpere. However this trend could be due to having more land (therefore more cities) around the equator with a wider range of humidity.
